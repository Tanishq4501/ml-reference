```python
import numpy as np
```

```python
# Define the 8 data points
points = {
    'P1': np.array([0.1, 0.6]),
    'P2': np.array([0.15, 0.71]),
    'P3': np.array([0.08, 0.9]),
    'P4': np.array([0.16, 0.85]),
    'P5': np.array([0.2, 0.3]),
    'P6': np.array([0.25, 0.5]),
    'P7': np.array([0.24, 0.1]),
    'P8': np.array([0.3, 0.2])
}

# Define initial centroids
m1 = points['P1']
m2 = points['P8']

print(f"Initial Centroid m1 (P1): {m1}")
print(f"Initial Centroid m2 (P8): {m2}")
```

```python
# Function to calculate Euclidean distance
def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))

# Initialize clusters
cluster1 = []  # Cluster for m1
cluster2 = []  # Cluster for m2
assignments = {} # To store which cluster each point belongs to

print("--- Cluster Assignment (Iteration 1) ---")

for name, point in points.items():
    # Calculate distance to both centroids
    dist_to_m1 = euclidean_distance(point, m1)
    dist_to_m2 = euclidean_distance(point, m2)
    
    # Assign to the closer cluster
    if dist_to_m1 <= dist_to_m2:
        cluster1.append(point)
        assignments[name] = 1
        print(f"Point {name}: dist to m1={dist_to_m1:.4f}, dist to m2={dist_to_m2:.4f} -> Assigned to Cluster 1")
    else:
        cluster2.append(point)
        assignments[name] = 2
        print(f"Point {name}: dist to m1={dist_to_m1:.4f}, dist to m2={dist_to_m2:.4f} -> Assigned to Cluster 2")

print("\n--- Answer 1 ---")
print(f"Q1: Which cluster does P6 belong to? A: Cluster {assignments['P6']}")
```

```python
# Convert lists of arrays to NumPy arrays for easier calculation
cluster1_points = np.array(cluster1)
cluster2_points = np.array(cluster2)

print("--- Answer 2 ---")
pop_m1 = len(cluster1_points)
pop_m2 = len(cluster2_points)
print(f"Q2: What is the population of a cluster around m2? A: {pop_m2}")
print(f"(Population of cluster around m1 is: {pop_m1})")
```

```python
# Calculate the new centroids by taking the mean of all points in each cluster
m1_updated = np.mean(cluster1_points, axis=0)
m2_updated = np.mean(cluster2_points, axis=0)

print("--- Answer 3 ---")
print(f"Q3: What is the updated value of m1 and m2?")
print(f"Updated m1 (mean of Cluster 1): {m1_updated}")
print(f"Updated m2 (mean of Cluster 2): {m2_updated}")
```

```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
# P1=[0.1,0.6], P2=[0.15,0.71], P3=[0.08,0.9], P4=[0.16, 0.85]
# P5=[0.2,0.3], P6=[0.25,0.5], P7=[0.24,0.1], P8=[0.3,0.2]

points = np.array([
    [0.1, 0.6],   # P1 (index 0)
    [0.15, 0.71], # P2 (index 1)
    [0.08, 0.9],  # P3 (index 2)
    [0.16, 0.85], # P4 (index 3)
    [0.2, 0.3],   # P5 (index 4)
    [0.25, 0.5],  # P6 (index 5)
    [0.24, 0.1],  # P7 (index 6)
    [0.3, 0.2]    # P8 (index 7)
])

point_names = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8']
```

```python
initial_centroids = np.array([
    [0.1,0.6],
    [0.3,0.2]
])
```

```python
kmeans = KMeans(n_clusters=2,init=initial_centroids,n_init=1,max_iter=1,random_state=42)
kmeans.fit(points)
```

```python
labels = kmeans.labels_+1
```
